```python
meshes = robot.collision_trimesh_fk().keys()
print(meshes)
```

```python
link_meshes = list(meshes)
print(link_meshes)
```

```python
num_links = len(link_meshes)
print(num_links)
```

```python
link_combos = list(itertools.combinations(range(len(meshes)), 2))
print(link_combos)
```

```python
collision_managers = []
for m in link_meshes:
    collision_manager = CollisionManager() # https://trimsh.org/trimesh.collision.html - collision check function
    collision_manager.add_object("link", m)
    collision_managers.append(collision_manager)
print(collision_managers)
```

```python
# Preprocess - find some collision-free configs and set min distances
batch_size = 250
rand_cfgs = (
    np.random.rand(batch_size, len(low_joint_vals))
    * (high_joint_vals - low_joint_vals)
    + low_joint_vals
)
print(rand_cfgs)
```

```python
mesh_poses = robot.collision_trimesh_fk_batch(rand_cfgs)
print(mesh_poses)
```

```python
colls = np.zeros(batch_size, dtype=np.bool)
print(colls)
```

```python
# link간의 거리 및 충돌 체크 함수
def check_pairwise_distances(mesh_poses, ind, boolean=False, normalize=False):
    if boolean:
        coll = False
    else:
        dists = np.zeros((num_links, num_links))
    for _, (i, j) in enumerate(link_combos):
        if abs(i - j) < 2 or ((i, j) == (6, 8)) or ((i, j) == (8, 10)):
            continue
        i_tf = mesh_poses[link_meshes[i]][ind]
        collision_managers[i].set_transform("link", i_tf)
        j_tf = mesh_poses[link_meshes[j]][ind]
        collision_managers[j].set_transform("link", j_tf)
        if boolean:
            coll |= collision_managers[i].in_collision_other(
                collision_managers[j]
            )
            if coll:
                return coll
        else:
            dists[i, j] = dists[j, i] = get_dist_pair(
                i, j, normalize=normalize
            )
    if boolean:
        return coll
    return dists

def get_dist_pair(self, i, j, normalize=False):
    min_dist = collision_managers[i].min_distance_other(
        collision_managers[j]
    )
    if normalize:
        min_dist = np.e ** (
            -2.0 * (min_dist / original_dists[i, j]) ** 2
        )
    return min_dist
```

```python
for k in range(batch_size):
    colls[k] = check_pairwise_distances(
        mesh_poses, k, boolean=True
    )
print(colls)
```

```python
original_dists = np.inf * np.ones(
    (num_links, num_links)
)
print(original_dists)
```

```python
for k, c in enumerate(colls):
    if not c:
        dists = check_pairwise_distances(
            mesh_poses, k, normalize=False
        )
        original_dists = np.minimum(original_dists, dists)
```
